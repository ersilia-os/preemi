```python
import pandas as pd
from pandas_profiling import ProfileReport


df = pd.read_csv("../data/processed/dr.csv")
df = df.drop(columns=["Method of Determining Gestation", "Last Menstrual Period", "Expected Due Date"])
profile = ProfileReport(df, title="PREEMI processed analysis")
profile.to_file("PREEMI_Processed_Analysis.html")
```

```python
window = 100
import numpy as np

y = list(df["Misscarriage"])
def proportion(y):
    y = np.array(y)
    v = list(set(y))[0]
    s = np.sum(y == v)
    p = s/len(y)
    if p > 0.5:
        return 1-p
    else:
        return p
```
